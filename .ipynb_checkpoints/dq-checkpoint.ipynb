```python
from pycbc import dq
import numpy as np
```

```python
#O3a start and end time
start_time = 1238166018
end_time = 1253977218
```

```python
# Get times that the Hanford detector has data
hsegs = dq.query_flag('H1', 'DATA', start_time, end_time)

# Get times that the Livingston detector has data
lsegs = dq.query_flag('L1', 'DATA', start_time, end_time)
```

```python
np.size(hsegs),np.size(lsegs)
```

```python
cotime=[]
```

```python
for i,[start_h,end_h] in enumerate(hsegs):
    for _,[start_l,end_l] in enumerate(lsegs):
        if start_h < end_l and end_h > start_l:
            cotime.append([max(start_h,start_l),min(end_h,end_l)])
```

```python
np.size(cotime)
```
